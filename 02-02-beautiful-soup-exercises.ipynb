```python
# imports
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re

# url to explore
url_medals = "https://es.wikipedia.org/wiki/Anexo:Medallero_de_los_Juegos_Ol%C3%ADmpicos_de_Tokio_2020#Medallas"

# get source code and check response status
response_medals = requests.get(url_medals)
print (f"The response to the request is {response_medals.status_code}")

# create a BeautifulSoup object to access the site content
soup_medals = BeautifulSoup (response_medals.content, "html.parser")
print (soup_medals.prettify())
```

```python
# find all the tables in the BeautifulSoup object and save them in a list
tables_list = soup_medals.find_all("table")
print("Showing table list ⬇️\n")
print (tables_list)
print(f"\nThe list has {len(tables_list)} tables.")
print("------------------------------------------\n")

# store the medals table (first table) in a variable 
medals_table = tables_list[0]
print("Showing medals table ⬇️\n")
print(medals_table)
print(f"\nData type: {type(medals_table)}")
print("------------------------------------------\n")

# find the table headers and store them in a variable
headers_list = medals_table.find_all("th")
print("Showing headers list ⬇️\n")
print(headers_list)
print(f"\nData type: {type(headers_list)}")
print("------------------------------------------\n")

# extract only headers text using list comprehansion
headers_text_list = [header.getText() for header in headers_list]
print(f"table headers are: {headers_text_list}")
print("------------------------------------------\n")

#remove the "\n" on each header
medals_table_headers = [header.replace("\n", "") for header in headers_text_list]
print(f"final medals table headers are: {medals_table_headers}")
print("------------------------------------------\n")
```

```python
# find all the rows within the medals table
rows_list = medals_table .find_all("tr")
print("Showing rows within medals table ⬇️\n")
print (rows_list)
print(f"\nThe list has {len(rows_list)} elements.")
print("------------------------------------------\n")

# extraxt text from each row
rows_clean = [row.text for row in rows_list]
print (f"The table has {len(rows_clean)}. See them listed below:")
print (rows_clean)
print("------------------------------------------\n")

# split the text gotten for each row using "\n". This will give us each cell value.

cells_values_list = [row.split("\n") for row in rows_clean] 
cells_values_list

```
